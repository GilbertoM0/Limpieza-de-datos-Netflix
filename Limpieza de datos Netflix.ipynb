## Columnas y descripciones:
- show_id: identificador único para cada programa (S1, S2).
- Tipo: especifica si el título es una 'película' o 'programa de televisión'.
- Título: El nombre del título de Netflix.
- Director: el director del título
- Elenco: los principales actores involucrados en el título.
- País: el país donde se produjo el título.
- date_added: la fecha en que se agregó el título a Netflix.
- Release_year: el año en que se lanzó originalmente el título.
-Calificación: la calificación de contenido ('PG-13', 'TV-Ma').
- Duración: duración de la película (en minutos) o el número de temporadas para programas de televisión.
- Listed_in: Categorías o géneros El título se encuentra en ('Documentos', 'Dramas de TV').
- Descripción: la descripción resumida

## Primero voy a rellenar los valores con 'desconocido' para no perder tantos datos con el metodo dropna de primeras

```python
# Rellenar valores faltantes en columnas categóricas
df['director'].fillna('Desconocido', inplace=True)
df['cast'].fillna('Desconocido', inplace=True)
df['country'].fillna('Desconocido', inplace=True)

df.dropna(inplace=True)
# Verificar si aún hay valores faltantes
df.info()
```

### El porque pienso que rellenar con desconocido es buena idea:
Al rellenar con "Desconocido", estoy conservando la gran mayoria de filas del dataset, lo que es útil si no quiero perder información valiosa en otras columnas. Eliminar filas con valores faltantes (dropna()) podría resultar en la pérdida de datos importantes, especialmente si los valores faltantes están concentrados en unas pocas columnas, por eso primero relleno con desconocido y despues ya puedo usar dropna sin perder tantos datos valiosos del dataset.

 ## Eliminar columnas no relevantes

## Limpiar y estandarizar texto
Eliminar espacios adicionales.

Convertir a minúsculas.


```python
# Limpiar y estandarizar texto
df['title'] = df['title'].str.strip().str.lower()
df['director'] = df['director'].str.strip().str.lower()
df['cast'] = df['cast'].str.strip().str.lower()
df['country'] = df['country'].str.strip().str.lower()
df['rating'] = df['rating'].str.strip().str.lower()
print(df['title'].unique())
print(df['director'].unique())
print(df['cast'].unique())
# print(df['country'].unique())
print(df['rating'].unique())
```

## Columna country

```python
# Y generemos las gráficas nuevamente
# Crear el gráfico de barras
# Crear una figura con subplots
fig, axes = plt.subplots(nrows=len(cols_categoricos), ncols=1, figsize=(10, 5 * len(cols_categoricos)))
fig.subplots_adjust(hspace=0.8)  # Ajustar el espacio entre subplots

# Bucle para graficar cada columna categórica
for i, col in enumerate(cols_categoricos):
    ax = axes[i]  # Seleccionar el subplot actual
    
    # Verificar si la columna tiene demasiados valores únicos
    if df[col].nunique() > 20:  # Si hay más de 20 valores únicos, graficar solo los 10 más frecuentes
        top_n = 10
        top_values = df[col].value_counts().nlargest(top_n).index
        sns.countplot(x=col, data=df[df[col].isin(top_values)], ax=ax, order=top_values, palette='Set2')
        ax.set_title(f'Top {top_n} {col} (de {df[col].nunique()} valores únicos)')
    else:
        # Graficar todos los valores si hay pocos valores únicos
        sns.countplot(x=col, data=df, ax=ax, palette='Set2')
        ax.set_title(f'Distribución de {col}')
    
    # Personalizar el subplot
    ax.set_xlabel('')
    ax.set_ylabel('Cantidad')
    ax.tick_params(axis='x', rotation=45, labelsize=8)  # Rotar etiquetas del eje x para mejor legibilidad

# Mostrar todos los gráficos
plt.show()
```
